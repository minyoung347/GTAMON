```python
# save adjacency matrix from keyword_list & keyword_similarity
# multiprocessing

import math
import numpy as np
from pprint import pprint
import gensim
import os
import csv
from multiprocessing import Pool

%cd Reddit

# hyper parameters of word2vec
eb_size = 50
w2v_win = 5
mincount = 100
epoch_size = 10

path = "model_save_{0}_{1}_{2}_{3}_recommend/keyword".format(eb_size,w2v_win,mincount,epoch_size)

def save_adj(month_start):
    for i in range(2012,2017):
        for iii in range(1,32):
            year = i
            month = month_start
            day_m = iii
            print(year,month,day_m)
            if os.path.isfile(path + "/keyword_{0}_{1}_{2}.csv".format(year,month,day_m)) == True:

                num_keyword = 0
                with open(path + "/keyword_{0}_{1}_{2}.csv".format(year,month,day_m), 'r') as f:
                    for line in f:
                        num_keyword += 1

                if not os.stat(path + "/keyword_sim_{0}_{1}_{2}.csv".format(year,month,day_m)).st_size == 0:
                    with open(path + "/keyword_sim_{0}_{1}_{2}.csv".format(year,month,day_m),'r') as f:
                        temp = []
                        rdr = csv.reader(f)
                        for line in rdr:
                            temp.append(line)
                    
                    adj_mat = [[0 for col in range(num_keyword)] for row in range(num_keyword)]
                    iteration = 0
                    for si in range(num_keyword):
                        for sii in range(num_keyword):
                            if int(temp[iteration][0]) == si and int(temp[iteration][1]) == sii:
                                adj_mat[si][sii] = 1
                                adj_mat[sii][si] = 1
                                if iteration < np.shape(temp)[0] -1 :
                                    iteration += 1
                                    
                    with open(path + "/adj_{0}_{1}_{2}.csv".format(year,month,day_m),"w+") as my_csv:
                        csvWriter = csv.writer(my_csv,delimiter=',')
                        csvWriter.writerows(adj_mat)

pool = Pool(processes = 12)
pool.map(save_adj,list(range(1,13)))
```
